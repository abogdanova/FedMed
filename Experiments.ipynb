All of the experiments were conducted on CIFAR10 dataset [ref]. It consists of 60000 small-resolution color images in 10 classes: airplane, automobile, bird, cat, deer, dog, frog, horse, ship, and truck; and is often used for benchmarking image classification algorithms. 
In order to test federated learning on semi-distributed data we assembled several training datasets. In all scenarios, we split 10000 images among 5 users and tested on a holdout data of 10000 images.
CIFAR IID: 5 users each hold 2000 images with even distribution of classes, so that each class is represented by 200 images
CIFAR non-IID: 5 users each hold 2000 images, but with 2 different classes missing in their set. 
CIFAR non-IID-extreme: 
The federated data set was composed from CIFAR10
